Et maintenant, l'analyse des correspondances multiples

```python
# Récupération des données

XLS = pd.read_csv('elza.csv',header=0,index_col=0)
# Affichage de la figure 1
print("\n", XLS,"\n")
col_names=(np.array(XLS.columns))
row_names=(np.array(XLS.index))
etiquettes = np.concatenate((row_names,col_names))
XLS = np.array(XLS)
print(XLS)
k = np.sum(XLS)
(n,p)=np.shape(XLS)
print(n,p)

# Création des matrices des poids, des métriques, et de toutes les matrices nécéssaires

DI=np.identity(n)
AJ=np.identity(n)
for i in range(n):
    sum = np.sum(XLS[i,:])
    DI[i,i] = sum/k
    AJ[i,i] = k/sum
# Affichage de la figure 2
print(pd.DataFrame(np.transpose(np.diagonal(DI)),index=row_names,columns=['marges']),"\n")

DJ=np.identity(p)
AI=np.identity(p)
for j in range(p):
    sum=np.sum(XLS[:,j])
    DJ[j,j] = sum/k
    AI[j,j] = k/sum
# Affichage de la figure 3
print(pd.DataFrame(np.transpose(np.diagonal(DJ)),index=col_names,columns=['marges']),"\n")

X = (1/k)*AJ@XLS
Y = (1/k)*AI@np.transpose(XLS)

VI = np.transpose(X)@DI@X
VJ = np.transpose(Y)@DJ@Y

# Analyse en composantes principales

vp,vep=np.linalg.eig(VI@AI)
vp = np.real(vp)
vep = np.real(vep)
vpb,vepb=np.linalg.eig(VJ@AJ)
vpb = np.real(vpb)
vepb = np.real(vepb)

XAU=X@AI@vep
YAV=Y@AJ@vepb

sum_vp = np.sum(vp)-1
inerties = (100/sum_vp)*vp[1:]

# Affichage de la figure 4
inertie = pd.DataFrame(np.transpose(inerties),columns=['inertie'])
print(inertie,"\n")
fig = px.bar(inertie, x=inertie.index, y='inertie')
fig.show()
cumulatif = np.transpose(inerties)
lengg = len(inerties)
print(lengg)
for i in range(1,lengg):
    cumulatif[i]+=cumulatif[i-1]


cumulatiff = pd.DataFrame(cumulatif,columns=['inertie'])

fig = px.bar(cumulatiff, x=inertie.index, y='inertie')
fig.show()
# Récupération des coordonnées dans le plan principal

vp1 = vp[1]
u1 = vep[:,1]
a1 = AI@u1
za1 = X@a1

vpb1 = vpb[1]
v1 = vepb[:,1]
b1 = AJ@v1
zb1 = Y@b1

vp2 = vp[2]
u2 = vep[:,2]
a2 = AI@u2
za2 = X@a2

vpb2 = vpb[2]
v2 = vepb[:,2]
b2 = AJ@v2
zb2 = Y@b2

vp3 = vp[3]
u3 = vep[:,3]
a3 = AI@u3
za3 = X@a3

vpb3 = vpb[3]
v3 = vepb[:,3]
b3 = AJ@v3
zb3 = Y@b3


coordx = np.concatenate((za1,zb1))
coordy = np.concatenate((za2,zb2))
coords = pd.DataFrame(np.transpose([coordx,coordy]),index=etiquettes,columns=['X1','X2'])
#Affichage de la figure 5
print(coords,"\n")

# Calcul des contributions

contribx1 = DI @ (np.square(za1))
contribx1 = (1/(np.sum(contribx1)))*contribx1
contribx2 = DI @ (np.square(za2))
contribx2 = (1/(np.sum(contribx2)))*contribx2
contriby1 = DJ @ (np.square(zb1))
contriby1 = (1/(np.sum(contriby1)))*contriby1
contriby2 = DJ @ (np.square(zb2))
contriby2 = (1/(np.sum(contriby2)))*contriby2
contrib1 = np.concatenate((contribx1,contriby1))
contrib2 = np.concatenate((contribx2,contriby2))
contrib = pd.DataFrame(np.transpose([contrib1,contrib2]),index=etiquettes,columns=['X1','X2'])
# Affichage de la figure 6
print(contrib,"\n")

# Calcul des cosinus carrés liés à la bonne représentation

represx1 = []
represx2 = []
for i in range(n):
    represx1.append((1/(np.sum(np.square(XAU[i,1:]))))*(za1[i]**2))
for i in range(n):
    represx2.append((1/(np.sum(np.square(XAU[i,1:]))))*(za2[i]**2))
represy1 = []
represy2 = []
for i in range(p):
    represy1.append((1/(np.sum(np.square(YAV[i,1:]))))*(zb1[i]**2))
for i in range(p):
    represy2.append((1/(np.sum(np.square(YAV[i,1:]))))*(zb2[i]**2))

repres1 = np.concatenate((represx1,represy1))
repres2 = np.concatenate((represx2,represy2))
# Affichage de la figure 7
print(pd.DataFrame(np.transpose([repres1,repres2]),index=etiquettes,columns=['X1','X2']),"\n")

# Graphique des modalités

# Affichage de la figure 8
fig = px.scatter(x=coordx, y=coordy, text=etiquettes,width=800, height=400,template="plotly_dark")
fig.update_traces(textposition='top center')
fig.show()
```

```python
# Parallèle avec les données supplémentaires

# Récupération des données

XLSp = pd.read_csv('beth.csv',header=0,index_col=0)
# Affichage de la figure 1
print("\n", XLSp,"\n")
col_namesp=(np.array(XLSp.columns))
row_namesp=(np.array(XLSp.index))
etiquettesp = np.concatenate((row_namesp,col_namesp))
XLSp = np.array(XLSp)
k = np.sum(XLSp)
(n,p)=np.shape(XLSp)
print(n,p)

# Création des matrices des poids, des métriques, et de toutes les matrices nécéssaires

DIp=np.identity(n)
AJp=np.identity(n)
for i in range(n):
    sum = np.sum(XLSp[i,:])
    DIp[i,i] = sum/k
    AJp[i,i] = k/sum
# Affichage de la figure 2
print(pd.DataFrame(np.transpose(np.diagonal(DIp)),index=row_namesp,columns=['marges']),"\n")

DJp=np.identity(p)
AIp=np.identity(p)
for j in range(p):
    sum=np.sum(XLSp[:,j])
    DJp[j,j] = sum/k
    AIp[j,j] = k/sum
# Affichage de la figure 3
print(pd.DataFrame(np.transpose(np.diagonal(DJp)),index=col_namesp,columns=['marges']),"\n")

Xp = (1/k)*AJp@XLSp
Yp = (1/k)*AIp@np.transpose(XLSp)

VIp = np.transpose(Xp)@DIp@Xp
VJp = np.transpose(Yp)@DJp@Yp

# Analyse en composantes principales

vp,vep=np.linalg.eig(VI@AI)
vp = np.real(vp)
vep = np.real(vep)
vpb,vepb=np.linalg.eig(VJ@AJ)
vpb = np.real(vpb)
vepb = np.real(vepb)

XAUp=Xp@AIp@vep
YAVp=Y@AJ@vepb

sum_vp = np.sum(vp)-1
inerties = (100/sum_vp)*vp[1:]

# Affichage de la figure 4
inertie = pd.DataFrame(np.transpose(inerties),columns=['inertie'])
print(inertie,"\n")
fig = px.bar(inertie, x=inertie.index, y='inertie')
fig.show()
cumulatif = np.transpose(inerties)
lengg = len(inerties)
print(lengg)
for i in range(1,lengg):
    cumulatif[i]+=cumulatif[i-1]


cumulatiff = pd.DataFrame(cumulatif,columns=['inertie'])

fig = px.bar(cumulatiff, x=inertie.index, y='inertie')
fig.show()
# Récupération des coordonnées dans le plan principal

vp1 = vp[1]
u1 = vep[:,1]
a1p = AIp@u1
za1p = Xp@a1p

vpb1 = vpb[1]
v1 = vepb[:,1]
b1p = AJ@v1
zb1p = Y@b1p

vp2 = vp[2]
u2 = vep[:,2]
a2p = AIp@u2
za2p = Xp@a2p

vpb2 = vpb[2]
v2 = vepb[:,2]
b2p = AJ@v2
zb2p = Y@b2p

vp3 = vp[3]
u3 = vep[:,3]
a3p = AIp@u3
za3p = Xp@a3p

vpb3 = vpb[3]
v3 = vepb[:,3]
b3p = AJ@v3
zb3p = Y@b3p


coordxp = np.concatenate((za1p,zb1p))
coordyp = np.concatenate((za2p,zb2p))
coordzp = np.concatenate((za3p,zb3p))
coordsp = pd.DataFrame(np.transpose([coordxp,coordyp,coordzp]),index=etiquettesp,columns=['X','Y','Z'])
#Affichage de la figure 5
print(coordsp,"\n")

# Calcul des contributions

contribx1p = DIp @ (np.square(za1p))
contribx1p = (1/(np.sum(contribx1p)))*contribx1p
contribx2p = DIp @ (np.square(za2p))
contribx2p = (1/(np.sum(contribx2p)))*contribx2p
contribx3p = DIp @ (np.square(za3p))
contribx3p = (1/(np.sum(contribx3p)))*contribx3p
contriby1p = DJp @ (np.square(zb1p))
contriby1p = (1/(np.sum(contriby1p)))*contriby1p
contriby2p = DJp @ (np.square(zb2p))
contriby2p = (1/(np.sum(contriby2p)))*contriby2p
contriby3p = DJp @ (np.square(zb3p))
contriby3p = (1/(np.sum(contriby3p)))*contriby3p
contrib1p = np.concatenate((contribx1p,contriby1p))
contrib2p = np.concatenate((contribx2p,contriby2p))
contrib3p = np.concatenate((contribx3p,contriby3p))
contribp = pd.DataFrame(np.transpose([contrib1p,contrib2p,contrib3p]),index=etiquettesp,columns=['X','Y','Z'])
# Affichage de la figure 6
print(contribp,"\n")

# Calcul des cosinus carrés liés à la bonne représentation

represx1p = []
represx2p = []
represx3p = []
for i in range(n):
    represx1p.append((1/(np.sum(np.square(XAUp[i,1:]))))*(za1p[i]**2))
for i in range(n):
    represx2p.append((1/(np.sum(np.square(XAUp[i,1:]))))*(za2p[i]**2))
for i in range(n):
    represx3p.append((1/(np.sum(np.square(XAUp[i,1:]))))*(za3p[i]**2))
represy1p = []
represy2p = []
represy3p = []
for i in range(p):
    represy1p.append((1/(np.sum(np.square(YAVp[i,1:]))))*(zb1p[i]**2))
for i in range(p):
    represy2p.append((1/(np.sum(np.square(YAVp[i,1:]))))*(zb2p[i]**2))
for i in range(p):
    represy3p.append((1/(np.sum(np.square(YAVp[i,1:]))))*(zb3p[i]**2))

repres1p = np.concatenate((represx1p,represy1p))
repres2p = np.concatenate((represx2p,represy2p))
repres3p = np.concatenate((represx3p,represy3p))
repres = pd.DataFrame(np.transpose([repres1p,repres2p,repres3p]),index=etiquettesp,columns=['X','Y','Z'])
# Affichage de la figure 7
print(repres,"\n")
```
